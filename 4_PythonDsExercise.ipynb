## 1. Write a NumPy program to get the indices of the sorted elements of a given array.

```python
import numpy as np
```

```python
score = np.array([103, 1502, 6230, 3, 1682, 5241, 4532])
print(score)
```

```python
def get_sorted_idx(np_array):
  array_sorted = np_array.copy()
  array_sorted.sort()
  return np.array([np.where(np_array == nr)[0][0] for nr in array_sorted])
```

```python
get_sorted_idx(score)
```

Alternatively there is the function "argsort()"

```python
np.argsort(score)
```
